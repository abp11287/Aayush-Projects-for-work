```python
import transformers
import torch

# Model location: https://huggingface.co/meta-llama/Meta-Llama-3.1-8B-Instruct
# Need huggingface

HF_TOKEN = "Put one here"

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    #model_kwargs={"torch_dtype": torch.float32},
    #model_kwargs={"torch_dtype": torch.float16},
    model_kwargs={"torch_dtype": torch.bfloat16},
    #model_kwargs={"torch_dtype": torch.int8},
    device_map="auto",
    token=HF_TOKEN,
)

```

# Ask Llama16 Function

```python
from IPython.display import display, Markdown
```

```python
def ask_llama_16(query):

    messages = [
        {"role": "system", "content": "You are a high school teacher."},
        #{"role": "user", "content": "Explain Maxwell's equations"},
        {"role": "user", "content": {query}},
    ]

    outputs = pipeline(
        messages,
        max_new_tokens=4000,
    )
    #print(outputs[0]["generated_text"][-1])

    answer = outputs[0]["generated_text"][-1]['content']

    return answer
```

```python
query = "Explain Maxwell's equations"
answer = ask_llama_16(query)

```

```python
display(Markdown(answer))
```

```python
def ask_openai_question(query):


    completion = client_openai.chat.completions.create(
    model="gpt-4o", #"model-identifier",
    messages=[
        {"role": "system", "content": "You are a high school teacher."},
        {"role": "user", "content": query}
    ],
    temperature=0.7,
    )

    #print(completion.choices[0].message)
    assessment = completion.choices[0].message.content

    print("Assessment: ", assessment)


    return assessment
```
